**compare the percentages**

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
# We can convert to a function, I am doing this in the last minute :( 

# Men watch more drama than women

movieratings = ratings.merge(movies, on = 'movie id')
totaldf = movieratings.merge(users, on = 'user id')
# Find the count of Drama watched by Men/Women
drama_movie_df1 = totaldf.groupby(by = ['gender','Drama']).count().reset_index()
drama_movie_df1

drama_movie_df2 = drama_movie_df1.groupby('gender').sum()
drama_movie_df2

drama_movie_df3 = drama_movie_df1.merge(drama_movie_df2, on='gender')
drama_movie_df3

# Finding the percentage based on each gender
drama_movie_df3['percent'] = drama_movie_df3['user id_x']/drama_movie_df3['user id_y'] * 100

drama_movie_df4 = drama_movie_df3[['gender','percent']]
drama_movie_df4
drama_movie_df4.sort_values('percent', ascending=False).head(1)

# 61.095997 Yes this is true
```
